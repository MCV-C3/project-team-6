```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import smoothings
```

```python
df = pd.read_csv("wandb_export_2025-12-13T17_52_09.707+01_00.csv")
```

```python
plt.figure(figsize=(14, 5))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', 
          '#9467bd', '#8c564b', '#e377c2']

widths = [pow(2, i) for i in range(5, 12)]

for width, color in zip(widths, colors):

    plt.plot(df['epoch'], df[f'Width only variation : {width} - test_accuracy'],  label=f"{width}", color=color)
    plt.plot(df['epoch'], df[f'Width only variation : {width} - train_accuracy'], linestyle="--", color=color)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
legend = plt.legend(by_label.values(), by_label.keys(), title="Run (--train / test)")
plt.setp(legend.get_title(), fontsize=9)
plt.title("Train and test accuracies for different widths")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.xlim(left=0)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
plt.show()
```
