```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("wandb_export_2025-12-13T17_52_09.707+01_00.csv")
```

```python
plt.figure(figsize=(14, 5))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', 
          '#9467bd', '#8c564b', '#e377c2']

widths = [pow(2, i) for i in range(5, 12)]

for width, color in zip(widths, colors):

    plt.plot(df['epoch'], df[f'Width only variation : {width} - test_accuracy'],  label=f"{width}", linestyle="--", color=color)
    plt.plot(df['epoch'], df[f'Width only variation : {width} - train_accuracy'],  label=f"{width}", color=color)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())
plt.title("Train and test accuracies for different widths")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.xlim(left=0)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
plt.show()
```
