```python
import wandb
from tqdm import tqdm

# Replace with your organization, project, and sweep ID
entity = "mcv-team-6"  # your org name
project = "C3-Week3"      # your project name
sweep_id = "k5ns8jzl"         # your sweep ID

api = wandb.Api()
sweep = api.sweep(f"{entity}/{project}/{sweep_id}")

best_acc = 0
best_run = None

for run in tqdm(sweep.runs, desc="Scanning sweep runs"):
    
    history = run.history(keys=["test_accuracy"])
    
    acc = 0
    if not history.empty:
        acc = history["test_accuracy"].max()
    
    if acc > best_acc:
        best_acc = acc
        best_run = run
        

print(f"Best test_accuracy: {best_acc}")
print("Configuration that achieved it:")
for k, v in best_run.config.items():
    print(f"  {k}: {v}")
```
